```python
import sqlite3, pandas as pd, textwrap, pathlib

sql_path = pathlib.Path(r"C:\Users\Owner\Documents\sql_samples_retail_analytics.sql")
with open(sql_path, "r", encoding="utf-8") as f:
    script = f.read()

con = sqlite3.connect(":memory:")
con.executescript(script)

def q(sql):
    return pd.read_sql_query(sql, con)

print("Rows:")
print(" products:", q("SELECT COUNT(*) AS n FROM products")["n"][0])
print(" customers:", q("SELECT COUNT(*) AS n FROM customers")["n"][0])
print(" orders:",    q("SELECT COUNT(*) AS n FROM orders")["n"][0])
print(" items:",     q("SELECT COUNT(*) AS n FROM order_items")["n"][0])

display(q("SELECT * FROM cohort_retention_12m LIMIT 40"))
display(q("SELECT * FROM rfm_scores ORDER BY r_score DESC, f_score DESC, m_score DESC LIMIT 20"))
display(q("SELECT * FROM category_margin_summary"))
con.close()
```
